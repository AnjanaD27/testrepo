```python
market_analysis=dataset.reindex(columns=['Order Date','Category'])
market_analysis.head()
```

```python
dataset['Order Date'] =pd.to_datetime(dataset['Order Date'])
dataset.head()
```

```python
da=market_analysis[market_analysis['Category']=='Furniture']
da.head()
```

```python
reorder_data=da.set_index('Order Date')
reorder_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
reorder_data.Category.resample('M').count().plot(kind='bar',figsize=(12,4),color='purple')
plt.ylabel('Furniture Sales')
plt.title('Furniture sales using OrderDates monthly wise')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
reorder_data.Category.resample('BQ').count().plot(kind='bar',color='red',figsize=(12,4))
plt.ylabel('Furniture Sales')
plt.title('Furniture sales using OrderDatesQuarterly wise')
plt.grid(True,color='b')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
reorder_data.Category.resample('A').count().plot(kind='bar',figsize=(12,4),color='c')
plt.ylabel('Furniture Sales')
plt.title('Furniture sales using OrderDates monthly wise')
plt.grid(True)
plt.legend()
plt.show()
```

```python
sal=dataset.reindex(columns=['Order Date','Sales'])
sal.head()
```

```python
sale=sal.set_index('Order Date')
sale.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
sales_rate = sale['Sales']

```

```python
sale.Sales.resample('M').mean().plot()
```

```python
#using additive model and seasonal_decompose 
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,4))
x=sale.Sales.resample('W').mean()
result = seasonal_decompose(x, model='additive')
#print(result.trend)
result.plot()
plt.show()
```

```python
q=dataset.reindex(columns=['Order Date','Quantity'])
q.head()
```

```python
quantity=q.set_index('Order Date')
quantity.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
quantity.Quantity.resample('M').mean().plot(figsize=(14,6))
plt.ylabel('Furniture Quantity')
plt.title('furniture quantity using order date(monthly).')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
quantity.Quantity.resample('W').mean().plot(color='purple',figsize=(14,6))
plt.ylabel('Furniture Quantity')
plt.title('furniture quantity using order date(weekly).')
plt.grid(True,color='black')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
quantity.Quantity.resample('A').mean().plot(color='blue',figsize=(14,6))
plt.ylabel('Furniture Quantity')
plt.title('furniture quantity using order date(yearly).')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
df_category = dataset[["Category", "Sales"]]
df_category.groupby('Category').sum().plot(kind="bar",color='red',figsize=(12,4))
plt.ylabel("sales")
plt.title('graph for sales according to catogery.')
plt.grid(True)
plt.legend()
plt.show()
```
